```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
```

```python
import findspark

findspark.init()
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
# cria df com os dados do censo da eduacação 2019
df_alunos = spark.read\
        .format('csv')\
        .option('header', True)\
        .option('inferSchema', True)\
        .option('delimiter', '|')\
        .load('../data/SUP_ALUNO_2019.CSV')

df_docentes = spark.read\
        .format('csv')\
        .option('header', True)\
        .option('inferSchema', True)\
        .option('delimiter', '|')\
        .load('../data/SUP_DOCENTE_2019.CSV')
```

```python
df_curso = spark.read\
        .format('csv')\
        .option('header', True)\
        .option('inferSchema', True)\
        .option('delimiter', '|')\
        .load('../data/SUP_CURSO_2019.CSV')
```

```python
# verificar schema do arquivo
df_alunos.printSchema()
```

```python
df_docentes.printSchema()
```

```python
df_curso.printSchema()
```

```python
# Verificando as primeiras linhas do df
df_alunos.limit(10).toPandas()
```

```python
# criando tempview para realizar analise dos daods
df_alunos.createOrReplaceTempView('CENSO')
df_docentes.createOrReplaceTempView('DOCENTE')
df_curso.createOrReplaceTempView('CURSO')
```

```python
# Questão 1
df_alunos.where(col('TP_COR_RACA') == 0).count()
```

```python
# Questao 2
df_alunos.where((col('TP_SEXO') == 1) & (col('CO_UF_NASCIMENTO') == 35)).count()
```

```python
# Questao 3
df_alunos.where((col('TP_SEXO') == 1) & (col('CO_UF_NASCIMENTO') == 43) & (col('IN_RESERVA_ETNICO') == 1)).count()
```

```python
# Questão 4
spark.sql('''
   SELECT CO_UF_NASCIMENTO, COUNT(1)

     FROM CENSO

    WHERE TP_SITUACAO IN (2, 5)

 GROUP BY CO_UF_NASCIMENTO

 ORDER BY 2 DESC
''').show()
```

```python
# Questão 5
spark.sql('''
   SELECT CO_UF_NASCIMENTO, COUNT(1)

     FROM CENSO

    WHERE IN_DEFICIENCIA_CEGUEIRA IN (1)

 GROUP BY CO_UF_NASCIMENTO

 ORDER BY 2 ASC
''').show()
```

```python
# Questão 6
spark.sql('''
   SELECT CO_UF_NASCIMENTO, AVG(NU_IDADE)

     FROM CENSO

    WHERE 1 = 1

 GROUP BY CO_UF_NASCIMENTO

 ORDER BY 2 DESC
''').show()
```

```python
# Questão 7
spark.sql('''
   SELECT CO_UF_NASCIMENTO, COUNT(1)

     FROM CENSO

    WHERE IN_BOLSA_PESQUISA IN (1)

 GROUP BY CO_UF_NASCIMENTO

 ORDER BY 2 DESC
''').show()
```

```python
# Questão 8
spark.sql('''
   SELECT COUNT(1)

     FROM CENSO

    WHERE TP_SEXO IN (1)
      AND TP_COR_RACA IN (2)
''').show()
```

```python
# Questão 9
spark.sql('''
   SELECT CO_UF_NASCIMENTO, COUNT(1)

     FROM DOCENTE

    WHERE TP_SEXO IN (1)

 GROUP BY CO_UF_NASCIMENTO

 ORDER BY 2 DESC
''').show()
```

```python
# Questão 10
spark.sql('''
   SELECT AVG(NU_IDADE)

     FROM DOCENTE

    WHERE TP_SEXO IN (2)
''').show()
```

```python
# Questão 11
spark.sql('''
   SELECT CO_UF_NASCIMENTO, COUNT(1)

     FROM DOCENTE

    WHERE TP_SEXO IN (2) AND TP_COR_RACA IN (5)

 GROUP BY CO_UF_NASCIMENTO

 ORDER BY 2 DESC
''').show()
```

```python
# Questão 11
spark.sql('''
   SELECT SUM(a.HOMENS_DOUTOR), SUM(a.MULHERES_DOUTOR),
         SUM(a.HOMENS_DOUTOR) - SUM(a.MULHERES_DOUTOR) as DIF
    FROM (
   SELECT  COUNT(1) HOMENS_DOUTOR, SUM(0) MULHERES_DOUTOR

     FROM DOCENTE

    WHERE TP_SEXO IN (1) AND TP_ESCOLARIDADE IN (5)
    UNION
    SELECT  SUM(0) HOMENS_DOUTOR, COUNT(1) MULHERES_DOUTOR

     FROM DOCENTE

    WHERE TP_SEXO IN (2) AND TP_ESCOLARIDADE IN (5)
    ) a 
''').show()
```

```python
# Questão 11
spark.sql('''
   SELECT COUNT(1) FROM CURSO WHERE NO_CURSO = 'DIREITO' AND CO_UF IN (21,22,23,24,25,26,27,28,29)
''').show()
```
